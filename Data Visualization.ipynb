```python
import matplotlib.pyplot as plt

prod = ['cosmetics', 'skincare', 'haircare']
q1 = [100, 200, 146]
q2 = [137, 260, 200]
q3 = [180, 172, 380]
q4 = [220, 240, 220]

q3_s = [q1[i] + q2[i] for i in range(3)]
q4_s = [q1[i] + q2[i] + q3[i] for i in range(3)]

plt.bar(prod,q1, width=0.3)
plt.bar(prod,q2, bottom = q1, width=0.3)
plt.bar(prod,q3, bottom = q3_s,width=0.3)
plt.bar(prod,q4, bottom = q4_s, width=0.3)

plt.show()
```

```python
# Q3.

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Sampledata.csv") 
grp = df.groupby('Product type')
mean = grp.mean()

prod_types = [key for key in grp.groups]
# print(prod_types)

plt.pie(mean['Price'], labels=prod_types)
plt.show()
```
